**Pre Processing**

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
import json

# Manual NLTK data path configuration
nltk.data.path.append("/usr/local/share/nltk_data")

# Load JSON data
with open('/content/bhasha-abhijnaanam.json', 'r') as f:
    json_data = json.load(f)
    data = json_data["data"]  # Access the data under the "data" key

# Initialize Snowball Stemmer for English
stemmer = SnowballStemmer("english")

# Preprocessing function
def preprocess(text):
    # Remove non-text characters
    text = re.sub(r"[^\w\s]", "", text)
    # Tokenization
    tokens = nltk.word_tokenize(text)
    # Remove stopwords
    tokens = [word for word in tokens if word.lower() not in stopwords.words('english')]
    # Stemming
    tokens = [stemmer.stem(word) for word in tokens]
    return tokens

# Preprocess each native sentence
for item in data:
    preprocessed_tokens = preprocess(item["native sentence"])
    print("Native Sentence:", item["native sentence"])
    print("Preprocessed Tokens:", preprocessed_tokens)
    print()
```

```python
# Initialize lists to store preprocessed tokens and language labels
X = []
y = []

# Preprocess each document and assign language labels
for item in data:
    preprocessed_tokens = preprocess(item["native sentence"])
    X.append(preprocessed_tokens)
    y.append(item["language"])
```

**Feature extraction and Train Test Split**

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.naive_bayes import MultinomialNB

# Flatten the list of preprocessed tokens into a list of strings
X_flattened = [' '.join(tokens) for tokens in X]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_flattened, y, test_size=0.2, random_state=42)

# Convert tokens into numerical features using TF-IDF
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Print the TF-IDF vectors for the first few samples in the test set
print("TF-IDF Vectors for the first 5 samples in the test set:")
for i in range(5):
    print(f"Sample {i+1}: {X_test_tfidf[i]}")
```

**Model Selection**

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.naive_bayes import MultinomialNB

'''# Train a Multinomial Naive Bayes Classifier
naive_bayes_classifier = MultinomialNB()
naive_bayes_classifier.fit(X_train_tfidf, y_train)

# Evaluate the model
y_pred_nb = naive_bayes_classifier.predict(X_test_tfidf)
print(classification_report(y_test, y_pred_nb))
'''
# Naive Bayes does not work accurately for large dataset as it considers independency between the events
# Whereas SVM works better than Naive Bayes but takes longer time to compute

# Train a Support Vector Classifier
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(X_train_tfidf, y_train)
```

**Model Evaluation**

```python
# Evaluate the model
y_pred = svm_classifier.predict(X_test_tfidf)
print(classification_report(y_test, y_pred))
```

**Pre Processing**

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Initialize Snowball Stemmer for English
stemmer = SnowballStemmer("english")

# Preprocessing function
def preprocess(text):
    # Tokenization
    tokens = nltk.word_tokenize(text)
    # Remove punctuation
    tokens = [re.sub(r'[^\w\s]', '', token) for token in tokens]
    # Remove stopwords
    tokens = [word for word in tokens if word.lower() not in stopwords.words('english')]
    # Stemming
    tokens = [stemmer.stem(word) for word in tokens]
    # Convert to lowercase
    tokens = [word.lower() for word in tokens]
    return tokens

# Preprocess each sentence in the DataFrame
df['Preprocessed'] = df['Sentences'].apply(preprocess)

# Display preprocessed DataFrame
print(df)
```

```python
import pandas as pd
import langid

def predict_language(sentence):
    langs = set()  # Use a set to store unique languages
    for word in sentence.split():
        lang = langid.classify(word)[0]  # Predict language using langid library
        langs.add(lang)
    return list(langs)  # Convert the set to a list

# Apply language prediction to each sentence
df['Language_Prediction'] = df['Sentences'].apply(predict_language)

# Mapping of language codes to names
language_names = {
    'kn': 'Kannada',
    'en': 'English',
    'hi': 'Hindi',
    'ne': 'Nepali',
    'ta': 'Tamil',
    'te': 'Telugu',
    'mr': 'Marathi',
    'bn': 'Bangla',
    'ml': 'Malayalam',
    'as': 'Assamese',
    'pa': 'Punjabi',
    'ur': 'Urdu',
    'gu': 'Gujarati',
    'sa': 'Sanskrit',
    'kok': 'Konkani',
    'mni': 'Manipuri',
    'brx': 'Bodo',
    'mai': 'Maithili',
    'doi': 'Dogri',
    'ks': 'Kashmiri',
    'or': 'Oriya'
}


# Replace language codes with names in Language_Prediction column
df['Language_Prediction'] = df['Language_Prediction'].apply(lambda langs: list(set([language_names[lang] if lang in language_names else lang for lang in langs])))

# Display the DataFrame with language names
print(df)
```
